```python
import numpy as np
import sklearn
import matplotlib
import matplotlib.pyplot as plt
import cv2
```

```python
def show_(img, h=10):  # show the image
    img = img.copy()
    img[img > 255] = 255
    img[img < 0] = 0
    img = img.astype(np.uint8)
    w = h
    plt.figure(figsize=(w, h))
    if img.ndim == 3:  # color image
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV 的色彩空間 BRG
        plt.imshow(img, vmin=0, vmax=255)
    else:  # gray image
        plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
img = cv2.imread('Hanne.jpg', 0)
show_(img, 5)
```

```python
ind = np.arange(img.size)
y = ind // img.shape[1]
x = ind % img.shape[1]
coor = np.vstack((y, x)).T
```

```python
p = (~img).flatten()
p = p / p.sum()
his = np.random.choice(np.arange(img.size), 1500000, p=p)
data = coor[his].T
data = data + np.random.uniform(size=data.shape)
```

```python
plt.figure(figsize=(16, 16))
plt.plot(data[1], -data[0], 'ko', ms=0.1)
plt.box(0)
plt.xticks([])
plt.yticks([])
# plt.savefig('res1.jpg', dpi=400)
```
