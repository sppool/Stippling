```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import cv2
```

```python
def show_(img, h=10):  # show the image
    img = img.copy()
    img[img > 255] = 255
    img[img < 0] = 0
    img = img.astype(np.uint8)
    w = h
    plt.figure(figsize=(w, h))
    if img.ndim == 3:  # color image
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV 的色彩空間 BRG
        plt.imshow(img, vmin=0, vmax=255)
    else:  # gray image
        plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
img = cv2.imread('Hanne.jpg', 0)
img = cv2.resize(img, None, fx=0.25, fy=0.25, interpolation=cv2.INTER_AREA)
show_(img)
img.shape
```

```python
n = 4
ar = ~img.flatten() // n
```

```python
ind = np.arange(img.size)
y = ind // img.shape[1]
x = ind % img.shape[1]
coor = np.vstack((y, x))
```

```python
plt.figure(figsize=(16, 16))
plt.axis('equal') 
plt.box(False)
plt.xticks([])
plt.yticks([])
for i in range(256 // n):
    plt.plot(coor[1][ind[ar == i]], -coor[0][ind[ar == i]], 'ko', ms=n * 0.012 * (i+1))

# plt.savefig('res2.jpg', dpi=400)
```
